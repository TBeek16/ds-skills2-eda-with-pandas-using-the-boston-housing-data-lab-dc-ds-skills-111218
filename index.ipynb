# Variable Descriptions

This data frame contains the following columns:

#### crim  
per capita crime rate by town.

#### zn  
proportion of residential land zoned for lots over 25,000 sq.ft.

#### indus  
proportion of non-retail business acres per town.

#### chas  
Charles River dummy variable (= 1 if tract bounds river; 0 otherwise).

#### nox  
nitrogen oxides concentration (parts per 10 million).

#### rm  
average number of rooms per dwelling.

#### age  
proportion of owner-occupied units built prior to 1940.

#### dis  
weighted mean of distances to five Boston employment centres.

#### rad  
index of accessibility to radial highways.

#### tax  
full-value property-tax rate per $10,000.

#### ptratio  
pupil-teacher ratio by town.

#### black  
1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town.

#### lstat  
lower status of the population (percent).

#### medv  
median value of owner-occupied homes in $10000s.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
# Use .describe() to return values for dispersion and centrality
df.describe()
```

```python
# Analyzing three variables
df["crim"].mean()
df["age"].quantile(.2)
df["indus"].cumsum()
```

```python
# Aim for large lots and regular lots
large_lots = df.loc[(df["zn"] > df.zn.quantile(0.75)),:]
reg_lots = df.loc[(df["zn"] <= df.zn.quantile(0.75)),:]
large_lots.describe()
```

```python
reg_lots.describe()
```

```python
# Look at buildings with more than 50% built before 1940
old_bldgs = df.loc[(df["age"] > 50),:]
new_bldgs = df.loc[(df["age"] <= 50),:]
```

```python
plt.figure()
plt.boxplot([old_bldgs.nox, new_bldgs.nox])
plt.xticks(ticks = [1,2], labels = ["Old", "New"])
plt.title("NOX in Old and New Buildings, Boston, MA")
plt.show()
```

```python
plt.figure()
plt.boxplot([old_bldgs["nox"], new_bldgs["nox"]])
plt.xticks(ticks = [1,2], labels = ["Old", "New"])
plt.title("NOX in Old and New Buildings, Boston, MA")
plt.show()
```

```python
# Define a new figure with matplotlib's .plot() function. 
fig = plt.figure()

# Add new axes to the figure with absolute positions
ax1 = fig.add_axes([0, 0, 1, 1]) # main axes
ax2 = fig.add_axes([0.7, 0.6, 0.3, 0.3])

ax1.hist(old_bldgs.nox)
ax1.set_title("Old Buildings")
ax1.set_xlabel("NOx")

ax2.hist(new_bldgs.nox)
ax2.set_title("New Buildings")
ax2.set_xlabel("NOx")

plt.show()
```

```python
# Define a new figure with matplotlib's .plot() function. 
fig = plt.figure()

# Add new axes to the figure with absolute positions
ax1 = fig.add_axes([0, 0, .4, .4]) # main axes
ax2 = fig.add_axes([0, 0.6, .4, .4])

ax1.hist(old_bldgs.nox)
ax1.set_title("Old Buildings")
ax1.set_xlabel("NOx")
# Set the limits of x and y for 1st plot 
ax1.set_xlim(0.3, 0.9), ax1.set_ylim(0, 50)

ax2.hist(new_bldgs.nox)
ax2.set_title("New Buildings")
ax2.set_xlabel("NOx")
# Set the limits of x and y for 2nd plot
ax2.set_xlim(0.3, 0.9), ax2.set_ylim(0, 50)

plt.show()
```

```python
# Let's see where distance value is first less than median, and then if it is greater than the median
near = df.loc[(df.dis <= df.dis.quantile(0.5)),]
far = df.loc[(df.dis > df.dis.quantile(0.5)),]

# Define a new figure with matplotlib's .plot() function. 
fig = plt.figure()

# Add new axes to the figure with absolute positions
ax1 = fig.add_axes([0, 0, .4, .4]) # main axes
ax2 = fig.add_axes([0, 0.6, .4, .4])

ax1.hist(near.nox)
ax1.set_title("Near Buildings")
ax1.set_xlabel("NOx")
# Set the limits of x and y for 1st plot 
ax1.set_xlim(0.3, 0.9), ax1.set_ylim(0, 50)

ax2.hist(far.nox)
ax2.set_title("Far Buildings")
ax2.set_xlabel("NOx")
# Set the limits of x and y for 2nd plot
ax2.set_xlim(0.3, 0.9), ax2.set_ylim(0, 50)

plt.show()
```

```python
# Define a new figure with matplotlib's .plot() function. 
figure = plt.figure(figsize = (10,6))

# Add new axes to the figure with absolute positions
ax1 = figure.add_subplot(121)
ax2 = figure.add_subplot(122)

ax1.scatter(x = near.nox, y = near.medv)
ax1.set_xlabel("Near")
ax2.scatter(x = far.nox, y = far.medv, color = 'orange')
ax2.set_xlabel("Far")

plt.show()
```

```python
# create a new column called Close
df.loc[df.dis > df.dis.quantile(0.75), 'close'] = 0
df.loc[df.dis <= df.dis.quantile(0.75), 'close'] = 1
df.head()

# Define a new figure with matplotlib's .plot() function. 
figure = plt.figure(figsize = (10,6))

# Add new axes to the figure with absolute positions
ax1 = figure.add_subplot(121)
ax2 = figure.add_subplot(122)

ax1.scatter(x = df.loc[df.close == 1, 'nox'], y = close.medv)
ax1.set_xlabel("Near")
ax2.scatter(x = far.nox, y = far.medv, color = 'orange')
ax2.set_xlabel("Far")

plt.show()
```
